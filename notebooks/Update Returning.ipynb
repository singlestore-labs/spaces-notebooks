# Update Returning

*_Update... Returning_* eliminates extra SELECT queries, reduces network round trips, and avoids race conditions between write and read. It simplifies application logic by making mutations atomic and observable, which is especially valuable for audit trails, event generation, and tightly coupled data pipelines.

You can skip the following SET GLOBAL if running on Helios. You can proceed with the rest of the notebook if "SELECT @@enable_dml_returning;" shows it's ON.



```python
%%sql
SET GLOBAL enable_dml_returning = ON; 
```

```python
%%sql
-- 1. Create a regular test table
CREATE TABLE update_returning_test_table (
    id SERIAL PRIMARY KEY,
    name TEXT,
    value INT
);
```

```python
%%sql
-- 2. Insert some test data
INSERT INTO update_returning_test_table (name, value) VALUES
('Alice', 10),
('Bob', 20),
('Charlie', 30),
('Diana', 40);
```

```python
%%sql
-- 3. Test UPDATE ... RETURNING
-- Increase value by 5 for all rows
UPDATE update_returning_test_table
SET value = value + 5
RETURNING id, name, value;
```

```python
%%sql
-- 4. Test DELETE ... RETURNING
-- Delete rows where value is greater than 30
DELETE FROM update_returning_test_table
WHERE value > 30
RETURNING id, name, value;
```

```python
%%sql
-- 5. Check remaining rows
SELECT * FROM update_returning_test_table;
```

```python
%%sql
-- 6. Clean up
DROP TABLE update_returning_test_table;
```
